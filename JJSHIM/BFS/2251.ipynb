[문제]

각각 부피가 A, B, C(1≤A, B, C≤200) 리터인 세 개의 물통이 있다.
처음에는 앞의 두 물통은 비어 있고, 세 번째 물통은 가득(C 리터) 차 있다.
어떤 물통에 들어있는 물을 다른 물통으로 쏟아 부을 때, 한 물통이 비거나 다른 한 물통이 가득 찰 때까지 물을 부을 수 있다. 이 과정에서 손실되는 물은 없다.
첫 번째 물통(용량이 A인)이 비어 있을 때, 세 번째 물통(용량이 C인)에 담겨있을 수 있는 물의 양을 모두 구한다.


입력
첫째 줄에 세 정수 A, B, C가 주어진다.

출력
첫째 줄에 공백으로 구분하여 답을 출력한다. 각 용량은 오름차순으로 정렬한다.

입력 예시: 8 9 10               출력 예시: 1 2 8 9 10

```python
import sys
from collections import deque

# x, y의 경우의 수 저장
def pour(x, y):                         #pour 함수는 물을 옮긴 후 새로운 상태가 방문되지 않았다면, 이를 방문한 것으로 표시하고 큐에 추가
    if not visited[x][y]:
        visited[x][y] = True
        q.append((x, y))

def bfs():          #bfs 함수는 너비 우선 탐색을 수행/ 큐가 비어있지 않은 동안 반복

    while q:
        # x : a물통의 물의 양, y : b물통의 물의 양, z : c물통의 물의 양
        x, y = q.popleft()
        z = c - x - y

        # 물통 A가 비어있을 때, 물통 C에 남아있는 물의 양을 answer 리스트에 추가
        if x == 0:
            answer.append(z)

        # x -> y
        water = min(x, b-y)
        pour(x - water, y + water)
        # x -> z
        water = min(x, c-z)
        pour(x - water, y)
        # y -> x
        water = min(y, a-x)
        pour(x + water, y - water)
        # y -> z
        water = min(y, c-z)
        pour(x, y - water)
        # z -> x
        water = min(z, a-x)
        pour(x + water, y)
        # z -> y
        water = min(z, b-y)
        pour(x, y + water)


# 입력(리터 범위)
a, b, c = map(int, sys.stdin.readline().split())    #sys.stdin.readline().split()을 통해 입력을 받고, map 함수를 사용하여 문자열을 정수로 변환

# 경우의 수를 담을 큐
q = deque()     #deque()는 큐를 생성. 큐는 너비 우선 탐색(BFS)을 위한 자료구조
q.append((0, 0))    #(0, 0)을 큐에 추가하여 처음에 두 물통이 비어있는 상태를 시작점으로 설정

# 방문 여부(visited[x][y])
visited = [[False] * (b+1) for _ in range(a+1)]     #visited는 방문 여부를 저장하는 2차원 리스트로, 물통 A와 B의 각 상태를 기록
visited[0][0] = True

# 답을 저장할 배열
answer = []

bfs()       #BFS 함수를 호출하여 탐색을 시작

# 출력
answer.sort()       #answer 리스트를 정렬하고 각 값을 출력
for i in answer:
    print(i, end=" ")


  # 탐색 중 A 물통이 비어있을 때마다 C 물통에 남아있는 물의 양을 기록
  # 물을 옮기는 다양한 경우를 처리하여 가능한 모든 상태를 탐색하는 방식으로 동작
```
